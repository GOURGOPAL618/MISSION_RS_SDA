# Time Series Analysis + Multi-temporal Imagery

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create daily NDVI (365 days)

days = np.arange(365)

# Sine wave peaking at day 200 (summer) with base NDVI 0.2 and peak 0.8
ndvi_daily = 0.5 + 0.3 * np.sin(2 * np.pi * (days - 110) / 365)
ndvi_daily += np.random.normal(0, 0.05, 365)
ndvi_daily = np.clip(ndvi_daily, 0.1, 0.9)

# 2. Resample to Monthly (12 values)

ndvi_monthly = ndvi_daily[:360].reshape(12, 30).mean(axis=1)

# 3. Plot
plt.style.use('dark_background')
plt.plot(ndvi_daily, label = 'DAILY NDVI', alpha = 0.4)
plt.plot(np.linspace(0, 365, 12), ndvi_monthly, 'ro-', label = 'Monthly Resampled')
plt.legend()
plt.tight_layout()
plt.show()
```
