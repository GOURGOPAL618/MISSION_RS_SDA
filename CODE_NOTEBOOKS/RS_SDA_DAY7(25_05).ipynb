# NDVI, NDWI, NDBI — Spectral Indices

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Create two numpy arrays for 100 pixels with realistic reflectance

np.random.seed(42)
B4_Red = np.random.uniform(0.05, 0.4, 100)
B8_NIR = np.random.uniform(0.05, 0.9, 100)

# 2.  Calculate NDVI

ndvi = (B8_NIR - B4_Red) / (B8_NIR + B4_Red)

# 3. Classify pixels based on threshold values

water = ndvi[ndvi < 0.0]
bare_soil = ndvi[(ndvi >= 0.0) & (ndvi <= 0.2)]
vagetation = ndvi[ndvi > 0.2]

print(f"Total Pixels: {len(ndvi)}")
print(f"Water Pixels (< 0.0): {len(water)}")
print(f"Bare Soil Pixels (0.0 - 0.2): {len(bare_soil)}")
print(f"Vagetation Pixels (> 0.2): {len(vagetation)}")
```

```python
# 4. Plot as histogram
plt.style.use('dark_background')
plt.figure(figsize = (10, 6))
plt.hist(ndvi, bins = 15, edgecolor = 'black', alpha = 0.7, color = 'grey', label = 'all Pixels')

# Visualizing threshold boundaries
plt.axvline(x = 0.0, color = 'blue', linestyle = '--', linewidth = 2, label = 'Water Boundary (0.0)')
plt.axvline(x = 0.2, color = 'brown', linestyle = '--', linewidth = 2, label = 'Soil Boundary (0.2)')
plt.title('DISTRIBUTION OF CALCULATED NDVI PIXELS', color = 'gold')
plt.xlabel('NDVI VALUE')
plt.ylabel('PIXEL COUNT')
plt.legend()
plt.grid(axis = 'y', alpha = 0.4)
plt.tight_layout()
plt.show()
```
