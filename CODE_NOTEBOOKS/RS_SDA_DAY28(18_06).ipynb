# SAR + Optical Data Fusion

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# 1. Generate Synthetic Data (1000 samples)
np.random.seed(42)
num_samples = 1000

# True labels (e.g., 0: Water, 1: Forest, 2: Urban, 3: Agriculture)
y = np.random.randint(0, 4, size = num_samples)

# SAR Features: VV, VH, and Ratio (VV/VH)
# Adding some relationship to y so the model can actually learn
sar_features = np.zeros((num_samples, 3))
sar_features[:, 0] = y * 2.0 + np.random.normal(0, 1.0, num_samples)    # VV
sar_features[:, 1] = y * 1.5 + np.random.normal(0, 0.8, num_samples)    # VH
sar_features[:, 2] = sar_features[:, 0] / (sar_features[:, 1] + 1e-5)   # Ratio
```

```python
# Optical Features: NDVI, NDWI, NDBI, EVI
optical_features = np.zeros((num_samples, 4))
optical_features[:, 0] = np.sin(y) + np.random.normal(0, 0.5, num_samples)  # NDVI
optical_features[:, 1] = np.cos(y) + np.random.normal(0, 0.5, num_samples)  # NDWI
optical_features[:, 2] = (y % 2) + np.random.normal(0, 0.6, num_samples)   # NDBI
optical_features[:, 3] = np.sin(y)*1.2 + np.random.normal(0, 0.4, num_samples) # EVI
```

```python
# 2. Concatenate into Fused Array (Feature-level fusion)
fused_features = np.concatenate((sar_features, optical_features), axis = 1)

print(f"SAR shape: {sar_features.shape}")
print(f"Optical shape: {optical_features.shape}")
print(f"Fused shape: {fused_features.shape}\n")
```

```python
# 3. Helper function to train and evaluate
def evaluate_model(X, y, dataset_name):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
    clf = RandomForestClassifier(n_estimators = 100, random_state = 42)
    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    acc = accuracy_score(y_test, preds)
    print(f"{dataset_name} Accuracy: {acc:.4f}")
    return acc
```

```python
# 4. Compare Models
sar_acc = evaluate_model(sar_features, y, "SAR Only")
opt_acc = evaluate_model(optical_features, y, "Optical Only")
fused_acc = evaluate_model(fused_features, y, "Fused Data")
```
