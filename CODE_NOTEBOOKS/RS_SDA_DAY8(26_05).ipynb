# Image Classification — Supervised vs Unsupervised

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# 1. Create a simulated remote sensing dataset (200 pixels, 3 features, 4 classes)

np.random.seed(42)
n_samples_per_class = 50

# Class 0: Water (High NDWI, Low NDVI, Low NDBI)

water = np.random.normal(loc = [-0.4, 0.6, -0.5], scale = 0.1, size = (n_samples_per_class, 3))

# Class 1: Vegetation (High NDVI, Low/Mid NDWI, Low NDBI)

vegetation = np.random.normal(loc = [0.7, -0.1, -0.3], scale = 0.1, size = (n_samples_per_class, 3))

# Class 2: Urban (High NDBI, Low NDVI, Low NDWI)

urban = np.random.normal(loc = [-0.1, -0.4, 0.5], scale = 0.1, size = (n_samples_per_class, 3))

# Class 3: Bare Soil (Mid NDVI, Low NDWI, Mid/High NDBI)

bare_soil = np.random.normal(loc = [0.2, -0.3, 0.2], scale = 0.1, size = (n_samples_per_class, 3))

# Combine into features (X) and labels (y)

X = np.vstack([water, vegetation, urban, bare_soil])
y = np.concatenate([np.zeros(50), np.ones(50), np.repeat(2, 50), np.repeat(3, 50)])
# Convert to a clean DataFrame for visualization

features = ['NDVI', 'NDWI', 'NVBI']
df = pd.DataFrame(X, columns = features)
df['class_label'] = y.astype(int)
```

```python
# 2. Split into Training and Testing Sets (80% train, 20% test)

X_train, X_test, y_train, y_test = train_test_split(df[features], df['class_label'], test_size=0.2, random_state=42, stratify=df['class_label'])

# 3. Train the Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 42)
rf_model.fit(X_train, y_train)

# 4. Predict and Evaluate
y_pred = rf_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_names = ['water', 'vegetation', 'urban', 'Bare soil']

print("--- EVALUATION RESULTS ---")
print(f"Overall Accuracy: {accuracy * 100:.2f}%\n")
print("Confusion Matrix:")
print(pd.DataFrame(conf_matrix, index=class_names, columns=class_names))
print("\nDetailed Classification Report:")
print(classification_report(y_test, y_pred, target_names=class_names))
```
