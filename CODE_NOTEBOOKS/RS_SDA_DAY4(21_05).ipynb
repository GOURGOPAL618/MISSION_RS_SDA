# Image Bands, Band Combinations + Spectral Signatures

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 13-band scale corresponding to Sentinel-2 Bands (B1 to B12, including B8A)
# Approximate wavelengths range from 443nm (B1) to 2190nm (B12)

bands = np.array(['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8', 'B8A', 'B9', 'B10', 'B11', 'B12'], dtype = str)

# 1. Clear Water: Peak in blue/green, plunges to near-zero in infrared

water = np.array([0.05, 0.06, 0.05, 0.03, 0.01, 0.005, 0.002, 0.001, 0.001,  0.001, 0.0, 0.0, 0.0])

# 2. Healthy Vegetation: Dip in red (B4), massive NIR spike (B8/8A), dips in SWIR water bands

vagetation = np.array([0.03, 0.04, 0.08, 0.03, 0.18, 0.35, 0.42, 0.45, 0.46, 0.40, 0.05, 0.22, 0.12])

# 3. Urban/Concrete: Highly reflective, relatively flat steady climb across all spectrums

urban = np.array([0.18, 0.20, 0.22, 0.24, 0.25, 0.26, 0.27, 0.28, 0.28, 0.25, 0.22, 0.30, 0.28])

# 4. Bare Soil: Smooth, steady upward slope from visible to shortwave infrared

bare_soil = np.array([0.08, 0.10, 0.14, 0.18, 0.22, 0.25, 0.28, 0.30, 0.31, 0.32, 0.33, 0.38, 0.40])
```

```python
# plotting configuration
plt.style.use('dark_background')
plt.figure(figsize = (16, 6))

plt.plot(bands, water, label = 'Clear Water', color = 'blue', marker = 'o', linewidth = 2)
plt.plot(bands, vagetation, label = 'Healthy Vegetation', color = 'green', marker = 's', linewidth = 2)
plt.plot(bands, urban, label = 'Urban / Concrete', color = 'gray', marker = '^', linewidth = 2)
plt.plot(bands, bare_soil, label = 'Bare Soil', color = 'peru', marker = 'd', linewidth = 2)

# Chart Formatting
plt.title("REALISTIC 13-BAND SPECTRAL SIGNATURES (SENTINEL - 2 SIMULATION)", fontsize = 14, fontweight = 'bold', color = 'gold')
plt.xlabel('Satellite Band Numbers', fontsize = 12)
plt.ylabel('Reflectance (0.0 - 1.0)', fontsize = 12)
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.ylim(-0.02, 0.55)
plt.legend(fontsize = 11, loc = 'upper left')

# Display the chart
plt.tight_layout()
plt.show()
```
