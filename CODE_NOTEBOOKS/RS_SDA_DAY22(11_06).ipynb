# SAR Fundamentals — Sentinel-1 Deep Dive

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 5 Land cover types: Water, Urban, Vegetation, Bare Soil, Flooded Area
classes = ['Water', 'Urban', 'Vegetattion', 'Bare soil', 'Flooded Area']

# Realistic mean dB values for Sentinel-1
# VV is generally stronger than VH

vv_means = np.array([-22.0, -5.0, -11.0, -14.0, -25.0])
vh_means = np.array([-28.0, -14.0, -17.0, -22.0, -30.0])

# To make it realistic, let's add some random radar speckle noise (variance)
vv_simulated = vv_means + np.random.normal(0, 1.0, len(classes))
vh_simulated = vh_means + np.random.normal(0, 1.0, len(classes))
```

```python
x = np.arange(len(classes))
width = 0.35

fig, ax = plt.subplots(figsize = (10, 6))
ax.bar(x - width/2, vv_simulated, width, label = 'vv (dB)', color = 'royalblue')
ax.bar(x + width/2, vh_simulated, width, label = 'vh (dB)', color = 'forestgreen')

ax.set_ylabel('Backscatter Intensity (dB)')
ax.set_title('SAR Backscatter (VV vs VH) across Land Cover Types')
ax.set_xticks(x)
ax.set_xticklabels(classes)
ax.legend()
ax.grid(axis = 'y', linestyle = '--', alpha = 0.8)
plt.tight_layout()
plt.show()
```
