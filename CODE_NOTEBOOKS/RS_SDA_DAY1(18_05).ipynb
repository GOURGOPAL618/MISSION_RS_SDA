#VISUALIZE A PIXXEL ARRAY


```python
import numpy as np
# Create a simple NumPy array of shape (13,) representing one pixel with 13 band values
pixel_data = np.array([10, 20, 25, 56, 48, 49, 18, 74, 95, 46, 95, 76, 85])

print("Numpy_array:")
print(pixel_data)
```

```python
import matplotlib.pyplot as plt

# visualize the array for the bar chart

plt.style.use('dark_background')
fig = plt.figure(figsize = (10, 6))
plt.bar(range(len(pixel_data)), pixel_data, color = '#D30000')
plt.xlabel('Band Index')
plt.ylabel('Band Value')
plt.title("PIXEL & BAND VALUES", color = 'gold')
plt.xticks(range(len(pixel_data)), [f'Band {i+1}' for i in range(len(pixel_data))], rotation = 45)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.8)
plt.tight_layout()
plt.show()
```
