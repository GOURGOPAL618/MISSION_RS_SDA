# REAL CROP MONITORING USING **NDVI** VALUES

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Create a dummay timline - series array of shape(36, 13)
# 36 steps (5-day intervals over 6 months), 13 bands matching Sentinel-2

np.random.seed(42)
dummy_data = np.random.uniform(0, 3000, size = (36, 13))  #Typical surface refectance values


# 2. Extract standard Sentinel-2 bands for NDVI calculation
# Band 4 (Red) is index 3, Band 8 (NIR) is index 7

b4_red = dummy_data[:, 3]
b8_nir = dummy_data[:, 7]

# 3. Calculate Normalized Difference Vegetation Index (NDVI)
# Formula: (NIR - Red) / (NIR + Red)

ndvi = (b8_nir - b4_red) / (b8_nir + b4_red)

# 4. Generate time steps representing every 5 days for 6 months
days = np.arange(0, 36 * 5, 5)

# 5. Plot the simulated NDVI profile over time
plt.style.use('dark_background')
plt.figure(figsize = (10, 5))
plt.plot(days, ndvi, marker = 'o', linestyle = '-', color = 'forestgreen', linewidth = 2)
plt.title('SIMULATED SENTINEL - 2 NDVI TIME SERIES FOR CROP MONITORING', color = 'gold')
plt.xlabel('Days From Start')
plt.ylabel('NDVI Value')
plt.grid(True, linestyle = '--', alpha = 0.8)
plt.ylim(-1, 1)
plt.tight_layout()
plt.show()
```
