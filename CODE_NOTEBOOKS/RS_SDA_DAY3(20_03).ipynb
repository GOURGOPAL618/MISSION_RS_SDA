```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Create a random NumPy array representing a Sentinel-2 image patch
# Shape: (Height=256 pixels, Width=256 pixels, Bands=13)
# Simulating 12-bit data scaled down or normalized reflectance

np.random.seed(42)
simulated_satellite_tile = np.random.uniform(0.0, 1.0, size=(256, 256, 13))

# 2. Extract Band 8 (NIR) and Band 4 (Red)
# Remember: Python uses 0-based indexing!
# Band 4 is at Index 3 | Band 8 is at Index 7

b4_red = simulated_satellite_tile[:, :, 3]
b8_nir = simulated_satellite_tile[:, :, 7]

# 3. Calculate NDVI manually using the formula: (NIR - Red) / (NIR + Red)
# Adding a tiny epsilon (1e-10) to prevent a RuntimeWarning if a denominator hits zero

ndvi = (b8_nir - b4_red) / (b8_nir + b4_red + 1e-10)
```

```python
# 4. Visualize the resulting matrix as a grayscale image

plt.style.use('dark_background')
plt.figure(figsize=(6, 6))
plt.imshow(ndvi, cmap='gray', vmin=-1, vmax=1)
plt.colorbar(label='NDVI Value Range')
plt.title('SIMULATED SATELLITE PATCH: MANUAL NDVI OUTPUT')
plt.axis('off')  # Hiding pixel coordinate axes for a clean map view
plt.show()
```
